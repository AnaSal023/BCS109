#Software Company

```python
package = 99
quantity = int(input('How many would you like to purchase?\n'))
discount = 0

if quantity >= 10 and quantity <= 19:
  discount = .1
elif quantity >= 20 and quantity <= 49:
  discount = .2
elif quantity >= 50 and quantity <100:
  discount = .3
elif quantity >= 100:
  discount = .4

total_amount = package*quantity*(1-discount)

print("The discount saved is ", discount*100, "%.")
print("The total amount of the current total is $", total_amount)
```

#Fast Freight Shipping Company

```python
package_weight = float (input('What is your package weight in pounds?\n'))
rate = 0

if package_weight <= 2:
  rate = 1.5
elif package_weight > 2 and package_weight <= 6:
  rate = 3.00
elif package_weight > 6 and package_weight <= 10:
  rate = 4.00
elif package_weight > 10:
  rate = 4.75

total_amount = float (package_weight*rate) 
print('The price of your package is $', total_amount)
```
