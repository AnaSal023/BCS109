##Create a function that takes in a item_cost, the percentage_off, and the tax.  It should return the total cost of the item.

```python
def total_cost(item_cost,percent_off,tax):
  percent_off = item_cost * percent_off
  item_cost = item_cost - percent_off
  total = item_cost + (item_cost * tax)
  print(total)

total_cost(140,.40,.10)  
```

##Create a function that takes in a number, and outputs all the numbers below it that are prime

```python
#prime number generator
def isPrime(num):
  for i in range(2,num):
    if (num % i == 0):
      return False
  return True 

def listPrimes(x,y):
  for i in range(x,y+1):
    if (isPrime(i)):
      print(i, end = " | ")

listPrimes(2,1000)
```

##Create a function that takes in a number and outputs all the numbers below it that are twin primes

```python
#twin prime
def isPrime(num):
  for i in range(2,num):
    if (num % i == 0):
      return False
  return True 

def is_twin_prime(x):
  if(isPrime(x) and isPrime(x+2)):
    return True
  return False 

def listTwinPrimes(x,y):
  for i in range(x,y+1):
    if (is_twin_prime(i)):
      print(i, " ", i+2)
      i += 2

listTwinPrimes(2,100)      
```

##Two people are playing the game rock, paper, scissors.  Create a function that finds the probability of winning for each option using 10,000 trials.  Output said probabilities.

```python
#rock,paper,scissors game
import random as r

def rps():
  rockCount = 0 
  paperCount = 0
  scissorsCount = 0
  for i in range(10000):
    num1 = r.randint(1,3)
    num2 = r.randint(1,3)

    if (num1 == num2):
      continue 
    elif (num1 == 1 and num2 == 2):
      paperCount +=1
    elif (num1 == 1 and num2 == 3):
      rockCount +=1
    elif (num1 == 2 and num2 == 3):
      scissorsCount +=1

  print("rock ", rockCount / 10000)
  print("scissors ", scissorsCount / 10000)          
  print("paper ", paperCount / 10000)

rps() 
```
