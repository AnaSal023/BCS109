1. Create functions to do the following: max, min, average, standard deviation, and geometric average. 



```python
import numpy as np
def theMax(numbers):
  my_max = numbers[0]
  for i in numbers:
    if my_max < i:
      my_max = i
  return my_max

def theMin(numbers):
  my_min = numbers[0]
  for i in numbers:
    if my_min > i:
      my_min = i
  return my_min

def theAverage(numbers):
  total = 0
  for i in numbers:
    total += i
  return total/len(numbers)  

def stdev(numbers):
  mean = sum(numbers) / len(numbers)
  variance = sum([((x - mean) ** 2) 
  for x in numbers]) / len(numbers)
  res = variance ** 0.5
  return res

def geometricAverage(numbers):
  mult = 1
  for i in numbers:
    mult *= i
  return pow(mult,1/len(numbers))

a = [8,6,2,9,13,3] 
print(theMax(a)) 
print(theMin(a))
print(theAverage(a))
print(stdev(a))
print(geometricAverage(a))
```

5. Create a function that returns a list of primes up until the number passed in. So if 100 is passed in, it should return all the primes up until 100.


```python
import numpy as np
def isPrime(num):
  for i in range(2,num):
    if (num % i == 0):
      return False
  return True 

def listPrimes(x,y):
  for i in range(x,y+1):
    if (isPrime(i)):
      print(i, end = "  ")

print(listPrimes(2,100))      
```

8. Find the sum of the even valued Fibonacci numbers below 500.


```python
def oddFib(k):
  first = 1
  second = 1
  total = first + second
  for i in range(k):
    third = first + second
    if (third % 2 == 1):
      total += third
    first = second
    second = third
  return total

print(oddFib(10))      
```

```python
import math as m
def fifth_powers():
  for spot in range(2,1000000):
    digits = [int(i) for i in str(spot)]
    total = 0
    for k in digits:
      total += m.pow(k,5)
    if total == spot:
      print(total)

fifth_powers()      
```
