```python
#This calculate the present value to see how much the user 
#should invest to obtain user's goal

#get desired future value, the annual interest rate, and number of years 
#that the money will grow and assing its own to a variable
future_value = float(input ("Enter the desired future value: "))

rate = float(input ("Enter the annual interest rate: "))

years = int(input ("Enter the number of years the money will grow: ")) 

#calculate the amount needed to deposit and assing it to 
#a present value variable
present_value = future_value / (1.0 + rate)**years

#display the amount needed to deposit
print ("You will need to deposit this amount: ", present_value)
```

```python
#This shows the user if he/she invest 5000 dollars, how much it will 
#grow at a 0.05% in 10 years

#assing amount to invest, the annual interest rate, and time in years in 
#its own variable 
present_value = 5000.0

rate = 0.05

years = 10

#calculate how much the money will grow 
future_value = present_value * (1 + rate)**years

#display the amount it will grow
print ("If you invest 5000.0 dollars at a 5%n rate" 
      + " in 10 years the money will grow: ", future_value)
```
