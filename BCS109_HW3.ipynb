#**Investing Futures**

###This will allow the investor to predict how much their invesment will grow at a certain rate and on a certain time

####Variables used to calculate present and future values:

*   future_value: the future value remaining after the final payment has been made
*   present_value: the current value of the annuity
*   years: period of time in which the value will grow
*   rate: the interest rate






###You can calculate how much you will need to invest to accomplish your goal
The formula used for this calculation is PV=FV/(1+i)n
  

*   PV = present value
*   FV = future value
*   i = interest rate
*   n = years





```python
#This calculate the present value to see how much the user 
#should invest to obtain user's goal

#get desired future value, the annual interest rate, and number of years 
#that the money will grow and assing its own to a variable
future_value = float(input ("Enter the desired future value: "))

rate = float(input ("Enter the annual interest rate: "))

years = int(input ("Enter the number of years the money will grow: ")) 

#calculate the amount needed to deposit and assing it to 
#a present value variable
present_value = future_value / (1.0 + rate)**years
present_value = round(present_value, 2)

#display the amount needed to deposit
print ("You will need to deposit this amount: $", present_value)
```

###This will show you a specific amount and how much it will grow on a specific time and at specific rate

The formula to calculate the future value is FV=PV(1+i)n

*   PV = present value
*   FV = future value
*   i = interest rate
*   n = years

```python
#This shows the user if he/she invest 5000 dollars, how much it will 
#grow at a 0.05% in 10 years

#assing amount to invest, the annual interest rate, and time in years in 
#its own variable 
present_value = 5000.0

rate = 0.05

years = 10

#calculate how much the money will grow 
future_value = present_value * (1 + rate)**years
future_value = round(future_value, 2)

#display the amount it will grow
print ("If you invest $5,000 at a 5% rate" 
      + " in 10 years the investment will grow to: $",future_value )
```
